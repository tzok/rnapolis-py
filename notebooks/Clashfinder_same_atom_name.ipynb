The 5G35 structure contains alternative atom positions
(with occupancy = 0.5), which can cause problems in many
bioinformatics tools.

The goal of this test is to compare the results of:
- the default run
- the run with the `--require-same-atom-name` option

### Default run

```python
!clashfinder \
  ../data/pdb/5g35.pdb \
  > ../outputs/5G35_default.txt
```

```python
!head ../outputs/5G35_default.txt
```

- 262 clashes found

### Run with the `--require-same-atom-name` option

```python
!clashfinder \
  --require-same-atom-name \
  ../data/pdb/5g35.pdb \
  > ../outputs/5G35_require_same_atom_name.txt
```

```python
!head ../outputs/5G35_require_same_atom_name.txt
```

- 100 clashes found

In the default mode, clashfinder detects significantly more clashes
than when the `--require-same-atom-name` option is enabled. This is
caused by the presence of alternative atom positions (occupancy = 0.5) 
in the 5G35 structure.

Without this option, the tool compares atoms from different alternative
conformations that do not occur simultaneously in reality, which leads
to the detection of false clashes.

The `--require-same-atom-name` option restricts comparisons to atoms
with the same name, resulting in fewer detected clashes and a more
realistic assessment of the structure.
